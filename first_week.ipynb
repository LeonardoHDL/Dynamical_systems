```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=math.pi/2

new_x=[]
for i in range(0,20):
    cx=math.cos(x)
    new_x.append(cx)
    x=new_x[i]
    
    

new_x.insert(0,math.pi/2)        
y=new_x
    
plt.plot(new_x)
plt.plot(new_x,y)
plt.show
```

```python
x=math.pi/4

new_x=[]
for i in range(0,20):
    cx=math.sin(x)
    new_x.append(cx)
    x=new_x[i]

new_x.insert(0,math.pi/4)    
y=new_x
    
plt.plot(new_x)
plt.plot(new_x,y)
plt.show
```
